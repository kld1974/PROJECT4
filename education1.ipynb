```python
#Reading data :
edu_df = pd.read_csv("Resources/df.csv")
```

```python
edu_df
```

```python
edu_clean_df = edu_df.rename(columns={"SATVR75" : "SAT Verbal Read 75",
                                     "SATMT75" : "SAT Math 75",
                                     "ACTEN75" : "ACT English 75",
                                     "ACTMT75" : "ACT Math 75"}
                                    )
edu_clean_df.columns
```

```python
edu_clean_df = edu_clean_df.fillna(0)
edu_clean_df=edu_clean_df.astype({'SAT Math 75': int,'SAT Verbal Read 75': int,'ACT English 75': int,'ACT Math 75': int})
edu_clean_df
```

```python
edu_clean_df = edu_clean_df[(edu_clean_df.Region != "Other US Jurisdctn") & (edu_clean_df.Region != "US Service Schools")]
edu_clean_df
```
